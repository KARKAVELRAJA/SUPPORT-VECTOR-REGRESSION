## ***IMPORTING LIBRARIES***

```python
import numpy as np
import pandas as pd
```

## ***LOADING THE DATASET***

```python
data = pd.read_csv("song_data.csv")
```

## ***SUMMARIZING THE DATASET***

```python
print(data.shape)
print(data.head())
```

## ***SEGREGATING THE DATASET INTO INPUT AND OUTPUT***

```python
input  = data.drop(['song_name','song_popularity','instrumentalness'],axis=1).values
print(input)
output = data.iloc[:,1:2].values
print(output)
```

## ***FEATURE SCALING***

```python
from sklearn.preprocessing import StandardScaler
scaling      = StandardScaler()
scaled_input = scaling.fit_transform(input)

print(scaled_input)
```

## ***SPLITTING THE DATASET INTO TRAIN AND TEST DATA***

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(scaled_input,output)
```

## ***TRAINING THE DATASET***

```python
from sklearn import svm

SVM_SVR_Model = svm.SVR(kernel="rbf")

SVM_SVR_Model.fit(x_train,y_train)
```

## ***PREDICTING THE RESULT USING THE TRAINED MODEL***

```python
y_pred = SVM_SVR_Model.predict(x_test)

y_pred
```
