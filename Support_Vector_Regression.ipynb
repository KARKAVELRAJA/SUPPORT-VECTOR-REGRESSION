## ***IMPORTING LIBRARIES***

```python
import numpy as np
import pandas as pd
```

## ***LOADING THE DATASET***

```python
data = pd.read_csv("Position_Salaries.csv")
```

## ***SUMMARIZING THE DATASET***

```python
print(data.shape)
print(data.head())
```

## ***SEGREGATING THE DATASET INTO INPUT AND OUTPUT***

```python
input  = data.iloc[:,1:2].values
print(input)
output = data.iloc[:,2:].values
print(output)
```

## ***FEATURE SCALING***

```python
from sklearn.preprocessing import StandardScaler
scaling_x       = StandardScaler()
scaling_y       = StandardScaler()
scaled_input    = scaling_x.fit_transform(input)
scaled_output   = scaling_y.fit_transform(output)


print(scaled_input)
print(scaled_output)
```

## ***VISUALIZING THE DATASET***

```python
import matplotlib.pyplot as plt
plt.xlabel("POSITION_LEVEL")
plt.ylabel("SALARY")
plt.scatter(scaled_input,scaled_output)
```

## ***TRAINING THE DATASET***

```python
from sklearn import svm

SVM_SVR_Model = svm.SVR(kernel="rbf")

SVM_SVR_Model.fit(scaled_input,scaled_output)
```

## ***CHECKING THE ACCURACY OF THE TRAINED MODEL***

```python
import matplotlib.pyplot as plt
plt.xlabel("POSITION_LEVEL")
plt.ylabel("SALARY")
plt.scatter(scaled_input,scaled_output)
plt.plot(scaled_input,SVM_SVR_Model.predict(scaled_input),color="Red")
```
