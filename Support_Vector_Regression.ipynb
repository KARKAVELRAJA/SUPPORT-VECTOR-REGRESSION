## ***IMPORTING LIBRARIES***

```python
import numpy as np
import pandas as pd
```

## ***LOADING THE DATASET***

```python
data = pd.read_csv("Age.csv")
```

## ***SUMMARIZING THE DATASET***

```python
print(data.shape)
print(data.head())
```

## ***SEGREGATING THE DATASET INTO INPUT AND OUTPUT***

```python
input  = data.iloc[:,0].values
input  = input.reshape(-1,1)
print(input)
output = data.iloc[:,-1].values
print(output)
```

## ***VISUALIZING THE DATASET***

```python
import matplotlib.pyplot as plt
plt.xlabel("AGE")
plt.ylabel("HEIGHT")
plt.scatter(input,output,color="green")
```

## ***SPLITTING THE DATASET INTO TRAIN AND TEST DATA***

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(input,output)
```

## ***TRAINING THE DATASET***

```python
from sklearn import svm

SVM_SVR_Model = svm.SVR(kernel="poly")

SVM_SVR_Model.fit(x_train,y_train)
```

## ***PREDICTING THE OUTPUT USING THE TRAINED MODEL***

```python
X = 10
X = np.reshape(X,(-1,1))

print(SVM_SVR_Model.predict(X),"IS THE HEIGHT OF THE PERSON WITH AGE",X)
```

## ***VISUALIZING THE RESULT OF THE TRAINED MODEL***

```python
import matplotlib.pyplot as plt


X_val = np.arange(min(x_train),max(x_train),0.01)
X_val = X_val.reshape(len(X_val),1)

plt.xlabel("AGE")
plt.ylabel("HEIGHT")
plt.scatter(x_train,y_train,color="green")

plt.plot(X_val,SVM_SVR_Model.predict(X_val),color="red")
```

## ***CALCULATING THE ACCURACY OF THE TRAINED MODEL***

```python
y_pred = SVM_SVR_Model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score, mean_squared_error

print("root_mean_square_error:",np.sqrt(mean_squared_error(y_test,y_pred)))
print("r2_score:",r2_score(y_test,y_pred)*100,"%")
```
